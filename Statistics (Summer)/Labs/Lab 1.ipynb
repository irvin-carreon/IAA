Import Packages

```python
import numpy as np 
import pandas as pd
import statsmodels as sm
from plotnine import *
```

Load Temperature Dataset

```python
temp = pd.read_csv("https://raw.githubusercontent.com/IAA-Faculty/statistical_foundations/refs/heads/master/normtemp.csv", sep =' ')
temp.head()
```

What is the mean of the temperature variable? (round to 2 decimal places)

```python
np.mean(temp['temperature'])
```

What is the minimum of the temperature variable? 

```python
np.min(temp["temperature"])
```

What is the maximum of the temperature variable?

```python
np.max(temp["temperature"])
```

What is the median of the temperature variable?

```python
np.median(temp['temperature'])
```

Create a QQ-plot for the temperature variable.

```python
p = (
    ggplot(temp, aes(sample="temperature")) +
    stat_qq() +
    stat_qq_line()
    )

p.show()
```

Create a boxplot for temperature with a reference line at 98.6 degrees. For a horizontal line: + geom_hline(yintercept = 98.6) For a vertical line: + geom_vline(xintercept = 98.6) 

```python
q = (
    ggplot(temp, aes(y="temperature")) +
    geom_boxplot() +
    geom_hline(yintercept = 98.6, color = "red")
)

q.show()
```

Using the Ameshousing dataset to answer the following questions using the SalePrice, Gr Liv Area, and log(SalePrice) variables.

```python
ames_py = pd.read_csv("https://raw.githubusercontent.com/IAA-Faculty/statistical_foundations/refs/heads/master/Ames.csv")
ames_py.head()
```

Create a histogram of SalePrice.

```python
r = (
    ggplot(ames_py, aes(x="Sale_Price")) +
    geom_histogram()
)

r.show()
```

Make a histogram of Log(Sale_Price).

```python
ames_py["logSale_Price"] = np.log(ames_py["Sale_Price"])

s = r = (
    ggplot(ames_py, aes(x="logSale_Price")) +
    geom_histogram(aes(y='..density..')) +
    geom_density()
)

r.show()
```

Make a histogram of 'Gr Liv Area' and overlay the kernel density estimator.

```python
t = (
ggplot(ames_py, aes(x='Gr_Liv_Area')) +
geom_histogram(aes(y='..density..'),
fill='lightgray', color='black') +
geom_density(color='black', size=1.5)
)

t.show()
```
